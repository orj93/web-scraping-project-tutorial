```python
from selenium import webdriver
import os
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.scrapethissite.com/pages/forms/")
```

```python
elements = driver.find_elements(By.TAG_NAME, 'table')
```

```python
for i in elements:
    print(i.text)
```

```python
string=elements[0].text
string
```

```python
lista=string.split('\n')
lista
```

```python
for i in range(len(lista)):
    lista[i]=lista[i].split(" ")
```

```python
matrix = [[float(j) for j in i[-7:]] for i in lista[1:]] 
matrix
```

```python
for i in lista[1:]:
    for j in i[-7:]:
        i.remove(j)
```

```python
teams=[" ".join(i) for i in lista[1:]]
teams
```

```python
for i in range(len(matrix)):
    matrix[i].insert(0,teams[i])
```

```python
matrix
```

```python
import pandas as pd
```

```python
df_columns = ['Team Name','Year','Wins','Loses','Win %','Goals for (GF)','Goals against (Ga)','+/-']
```

```python
df = pd.DataFrame(matrix,columns = df_columns)
df
```

```python
final_df = df.sort_values(by='Wins', ascending=False)
final_df
```

```python
import matplotlib.pyplot as plt
```

```python
labels = final_df['Team Name']
values = final_df['Wins']

plt.figure(figsize = (15, 10))

plt.bar(labels, values)
plt.xticks(rotation=90)
plt.title("Wins of hockey teams")
plt.show()
```
